### date and time

```python
import datetime
```

```python
# current datetime
now = datetime.datetime.now()
print(now)
```

```python
# date
date = datetime.date(2025,4,1)
print(date)
```

```python
# time
time = datetime.time(15,15,15)
print(time)
```

```python
# date and time
date_time = datetime.datetime(2024,1,4,15,15,15)
print(date_time)
```

```python
# split
now = datetime.datetime.now()
year = now.year
month = now.month
day = now.day

print(year)
print(month)
print(day)
```

### Regular Expressions

```python
import re
```

```python
text = "Mi gato esta en el tejado y mi otro gato esta en el jardin"
```

```python
matches = re.findall("gato", text)
matches
```

```python
text_2 = "Hoy estamos a 18/11/2025, mañana estaremos a 19/11/2025"

matches_2 = re.findall("\d{2}\/\d{2}\/\d{4}", text_2)

print(matches_2)
```

```python
text_3 = "Los usuarios pueden contactarnos a soporte@hack4u.io o a info@hack4u.io"

matches_3 = re.findall("(\w+)@(\w+\.\w{2,})", text_3)
print(matches_3)
```

```python
# sub()
text_4 = "MI gato esta ene l tejado y mi otro gato esta en el jardin"

new_text = re.sub("gato", "perro", text_4)
print(new_text)
```

```python
# split()
text_5 = "campo1, campo2, campo3"
split_text = re.split(",", text_5)
print(split_text)
```

### Files and Dirs

```python
import os
```

```python
# file in current dir
if os.path.exists("cac_config.txt"):
    print(f"\n[+] File exists")
else:
    print(f"\n[-] File does not exist")
```

```python
# dirs
if not os.path.exists("cacs_dir"):
    os.mkdir("cacs_dir")
```

```python
# subdirectories
if not os.path.exists("my_dir/my_subdir"):
    os.makedirs("my_dir/my_subdir")
```

```python
# for i in $(seq 1 5); do touch file$i.txt; done
print(f"\n[+] Listando archivos:\n")
rec = os.listdir()
for file in rec:
    print(file)
```

```python
# remove
if os.path.exists("file.txt"):
    os.remove("file.txt")
```

```python
# remove dir
if os.path.exists("/tmp/test"):
    os.rmdir("/tmp/test") # only in empty dir
    print("Directory removed")
```
